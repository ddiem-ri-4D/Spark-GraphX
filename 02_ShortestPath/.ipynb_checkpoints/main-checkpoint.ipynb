# Demo thuật toán tìm đường đi ngắn nhất trên đồ thị vô hướng có trọng số bằng thuật toán BFS.
* GraphFrame không hỗ trợ tìm đường đi ngắn nhất trên đồ thị có trọng số, để làm được điều này chúng ta cần custom lại vài thứ.

```python
from graphframes.lib import AggregateMessages as AM
from pyspark.sql import functions as F
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark import SparkContext
from graphframes import *
import matplotlib.pyplot as plt
import networkx as nx
import findspark
import pandas as pd

findspark.init()
```

```python
spark = SparkSession.builder.appName("Shortest_path").getOrCreate()
sc = spark.sparkContext
```

* Đầu tiên, chúng ta cần định nghĩa một hàm _(user defined function)_ mà hàm này sẽ cho chúng ta biết được đường đi từ **source** cho đến **destination**. Đây chính là đường đi ngắn nhất mà ta tìm được.

```python
def createTransportGraph():
    # định nghĩa một node bao gồm các trường dữ liệu gì
    node_fields = [
        StructField("id", StringType(), True),
        StructField("latitude", FloatType(), True),
        StructField("longitude", FloatType(), True),
        StructField("population", IntegerType(), True)
    ]
    
    # load dữ liệu chứa các thông tin node vào @nodes
    nodes = spark.read.csv("./data/transport-nodes.csv", 
                           header=True, 
                           schema=StructType(node_fields))
    
    # load dữ liệu chứa các thông tin về relationship vào @rels
    rels = spark.read.csv("./data/transport-relationships.csv",
                          header=True)
    
    """
    Vì bài toán của chúng ta là trên đồ thị vô hướng => ta cần thêm một đường
    đi ngược lại từ destination đến source.
    """
    reversed_rels = rels.withColumn("newSrc", rels.dst).\
                        withColumn("newDst", rels.src).\
                        drop("src", "dst").\
                        withColumnRenamed("newSrc", "src").\
                        withColumnRenamed("newDst", "dst").\
                        select("src", "dst", "relationship", "cost")
                        
    """
    Cuối cùng, ta cần kết hợp @rels và @reversed_rels lại thàng một
    dữ liệu chứa các cạnh duy nhất và lưu vào @relationships.
    """
    relationships = rels.union(reversed_rels)
    
    # trả về một GraphFrame
    return GraphFrame(nodes, relationships)
```

```python
g = createTransportGraph()
```

* Tiếp theo, chúng ta sẽ viết hàm tìm đường đi ngắn nhất như sau:

```python
add_path_udf = F.udf(lambda path, id: path + [id], ArrayType(StringType())) 
```

```python
def shortestPath(g, origin, destination, column_name="cost"):
    # nếu trong graph không tồn tại đỉnh nào có id bằng @destination => thuật toán không cần chạy
    if g.vertices.filter(g.vertices.id == destination).count() == 0:
        return (spark.createDataFrame(sc.emptyRDD(), g.vertices.schema).withColumn("path", F.array()))

    """
    Tạo ra những đỉnh kèm theo thuộc tính và lưu vào biến @vertices có kiểu là Spark-DataFrame với các cột:
        - visited: lưu trạng thái đã đi qua hay chưa, khởi tạo ban đầu toàn bộ là False bằng hàm `F.lit(False)`
        - distance: lưu chi phí của đường đi ngắn nhất đến điểm đó, ở đây mọi điểm đều là infinity trừ đỉnh gốc đi
            đến chính nó có chi phí bằng 0
        - path: mảng dùng để lưu vết đường đi ngắn nhất, mặc định ban đầu ở mọi đỉnh đều là rỗng
    """
    vertices = (g.vertices.withColumn("visited", F.lit(False))
                .withColumn("distance", F.when(g.vertices["id"] == origin, 0)
                            .otherwise(float("inf")))
                .withColumn("path", F.array()))
    
    """
    Copy toàn bộ Spark-DataFrames @vertices và lưu vào vùng nhớ cache nhằm tối ưu hiệu năng truy xuất,
    sau đó trả về và lưu tham chiếu vào biến @cache_vertices, biến này sẽ truy xuất thường xuyên nên cần
    lưu vào vùng nhớ đệm
    """
    cached_vertices = AM.getCachedDataFrame(vertices)
    
    """
    Tạo một graph mới với các đỉnh tham chiếu đến vùng nhớ cache @cached_vertices, các cạnh
    tham chiếu đến các cạnh trong @g (tức graph input ban đầu)
    """
    g2 = GraphFrame(cached_vertices, g.edges)
    
    # khi vẫn còn đỉnh chưa ghé thăm
    while g2.vertices.filter('visited == False').first():
        # lấy id của đinh chưa ghé thăm có @distance nhỏ nhất, khi mới run lần đầu tiên là id của đỉnh gốc
        current_node_id = g2.vertices.filter('visited == False').sort("distance").first().id
        
        """
        Tạo một "message" để tính chi phí đuòng đi mới, việc này giống như ta tính tổng hai biến interger,
        nhưng do ta đang làm việc trên Spark-DataFrame, mọi thứ đều được tính toán trên các column của 
        Spark-DataFrame, nên ta cần chuẩn bị một thứ gọi là "message" để tính toán
        
        Ở đây, chi phí đường đi bằng cách lấy chi phí mà ta lưu lại trên đỉnh cộng
        cho giá trị `cost` của cạnh nối giữa 2 đỉnh
        """
        msg_distance = AM.edge[column_name] + AM.src['distance']
        
        """
        Tương tự, ta cũng cần định nghĩa một "message" để lấy đường đi mới, nhưng nó sẽ là một mảng
        
        Ở đây, đường đi mới là đường đi trước đó đến @current_node_id + id của đỉnh hiện tại, đường
        đi mới này sẽ được lưu trên node hàng xóm
        """
        msg_path = add_path_udf(AM.src["path"], AM.src["id"])
        
        """
        Ở đây, ta chuẩn bị một "message", nếu đỉnh hiện tại là đỉnh đang xét => ta sẽ chuyền message này đến
            các đỉnh hàng xóm của đỉnh này, ngược lại "message" này là null (mặc định)
        
        "Message" chứa giá trị do ta định nghĩa là một kiểu dữ liệu mới gồm 2 thành phần:
            - thành phần đầu tiên là numeric - là biến @msg_distance, chi phí đường đi đến 1 đỉnh hàng xóm
            - thành phần thứ hai là mảng - là biến @msg_path chính là đường đi trước đó để đến được đỉnh hàng xóm
            * Ta đóng gói mọi thứ được đóng gói trong hàm `F.struct`
        """
        msg_for_dst = F.when(AM.src['id'] == current_node_id, F.struct(msg_distance, msg_path))
        
        """
        Tiếp theo, ta gửi message này đến các đỉnh hàng xóm bằng hàm `aggregateMessages`, kết quả trả về
        ta chỉ lấy đỉnh hàng xóm có giá trị có distance nhỏ nhất, ta lưu lại vào cột `aggMess`.
        
        Ở đây, vì sao ta cần sài hàm `F.min()`, vì 1 đỉnh sẽ có thể có nhiều đình khác có thể đi đến đỉnh đó,
        nền ta cần chọn ra đỉnh có chi phí thấp nhất.
        """
        new_distances = g2.aggregateMessages(F.min(AM.msg).alias("aggMess"), sendToDst=msg_for_dst)
        
        """
        Tiếp theo, ta duyệt lại Spark-DataFrame, các đỉnh nào trước đây đã được thăm và đỉnh hiện tại sẽ được đánh dấu
        là true, ngược lại false.
        """
        new_visited_col = F.when(g2.vertices.visited | (g2.vertices.id == current_node_id), True).otherwise(False)
        
        """
        Tiến hành cập nhật chi phí đường đi nhỏ nhất cho các đỉnh lân cận của @current_node_id và đường đi nếu chi phí di chuyển
        có giảm, ở đây `col1` chính là chi phí đường đi, `col2` là mảng lưu vết đường đi
        """
        new_distance_col = F.when(new_distances["aggMess"].isNotNull() & (new_distances.aggMess["col1"] < g2.vertices.distance), new_distances.aggMess["col1"]).otherwise(g2.vertices.distance)
        new_path_col = F.when(new_distances["aggMess"].isNotNull() & (new_distances.aggMess["col1"] < g2.vertices.distance), new_distances.aggMess["col2"].cast("array<string>")).otherwise(g2.vertices.path)
        new_vertices = (g2.vertices.join(new_distances, on="id",how="left_outer")
                        .drop(new_distances["id"])
                        .withColumn("visited", new_visited_col)
                        .withColumn("newDistance", new_distance_col)
                        .withColumn("newPath", new_path_col)
                        .drop("aggMess", "distance", "path")
                        .withColumnRenamed('newDistance', 'distance')
                        .withColumnRenamed('newPath', 'path'))
        
        """
        Sau khi cập nhập lại các thuộc tính cho đỉnh, ta tiếp tục tao Spark-DataFrame mới và lưu vào cache
        với @cached_new_vertices được tham chiếu đến
        """
        cached_new_vertices = AM.getCachedDataFrame(new_vertices)
        g2 = GraphFrame(cached_new_vertices, g2.edges) # tạo lại GraphFrame cho @g2 với đồ thị đạ được cập nhật
        
        """
        Kiểm tra đã đến đích chưa, nếu đã đến dừng thuật toán, trả về Spark-DataFrame kết quả
        """
        if g2.vertices.filter(g2.vertices.id == destination).first().visited:
            return (g2.vertices.filter(g2.vertices.id == destination)
                    .withColumn("newPath", add_path_udf("path", "id"))
                    .drop("visited", "path")
                    .withColumnRenamed("newPath", "path"))

    """
    Ngược lại, ta gọi đệ quy để tìm đường đi nhỏ nhất tới đích
    """
    return (spark.createDataFrame(sc.emptyRDD(), g.vertices.schema)
            .withColumn("path", F.array()))
```

```python
res = shortestPath(g, "Amsterdam", "Colchester", "cost")
res.select("id", "distance", "path").show(truncate=False)
```

> Nhận xét:
> * Trong khi với hàm tìm kiếm đường đi `bfs()` do Spark cung cấp, tốc độ giải quyết bài toán chỉ trong khoảng dưới 3s.
> * Với hàm custom này do chúng ta thực hiện, performace không tốt, tận hơn 40s $\Rightarrow$ chỉ phù hợp với việc nghiên cứu, thử nghiệm hơn là áp dụng vào một tập dữ liệu lớn đích thực.

```python
def plotGraph(data_path, path=None):
    data = pd.read_csv(data_path)
    edge_labels = {}
    plt.figure(figsize=(12, 8))
    Gplot=nx.Graph()

    step = []
    if path:
        for i in range(1, len(path)):
            step.append((path[i-1], path[i]))
            step.append((path[i], path[i - 1]))
    
    for _, row in data.iterrows():
        if (row['src'], row['dst']) in step:
            Gplot.add_edge(row['src'], row['dst'], weight=4, color="red")
        else:
            Gplot.add_edge(row['src'], row['dst'],  weight=2,  color="black")
        edge_labels[(row['src'], row['dst'])] = row['cost']
        
    colors = nx.get_edge_attributes(Gplot,'color').values()
    weights = nx.get_edge_attributes(Gplot,'weight').values()

    
    pos = nx.spring_layout(Gplot)
    nx.draw(Gplot, pos, with_labels=True, font_weight='bold', edge_color=colors, width=list(weights))
    nx.draw_networkx_edge_labels(Gplot, pos, edge_labels=edge_labels, font_color='blue')
```

```python
plotGraph("./data/transport-relationships.csv", res.select("path").take(1)[0]['path'])
```
