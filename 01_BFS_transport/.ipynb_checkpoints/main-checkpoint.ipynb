* Ta có thể trực quan hóa do data này đơn giản.

```python
def plotGraph(data_path, path=None):
    data = pd.read_csv(data_path)
    edge_labels = {}
    plt.figure(figsize=(12, 8))
    Gplot=nx.Graph()

    step = []
    if path:
        for i in range(1, len(path)):
            step.append((path[i-1], path[i]))
            step.append((path[i], path[i - 1]))
    
    for _, row in data.iterrows():
        if (row['src'], row['dst']) in step:
            Gplot.add_edge(row['src'], row['dst'], weight=4, color="red")
        else:
            Gplot.add_edge(row['src'], row['dst'],  weight=2,  color="black")
        edge_labels[(row['src'], row['dst'])] = row['cost']
        
    colors = nx.get_edge_attributes(Gplot,'color').values()
    weights = nx.get_edge_attributes(Gplot,'weight').values()

    
    pos = nx.spring_layout(Gplot)
    nx.draw(Gplot, pos, with_labels=True, font_weight='bold', edge_color=colors, width=list(weights))
    nx.draw_networkx_edge_labels(Gplot, pos, edge_labels=edge_labels, font_color='blue')
```

```python
plotGraph("./data/transport-relationships.csv")
```

* Tiếp theo, ta sẽ import một vài thư viện cũng như khởi tạo **SparkSession**.

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from graphframes import *
import findspark

findspark.init()
```

```python
spark = SparkSession.builder.appName("BFS-transport").getOrCreate()
```

* Tiếp theo, tạo **GraphFrames** từ file ***.csv**.

```python
def createTransportGraph():
    # định nghĩa một node bao gồm các trường dữ liệu gì
    node_fields = [
        StructField("id", StringType(), True),
        StructField("latitude", FloatType(), True),
        StructField("longitude", FloatType(), True),
        StructField("population", IntegerType(), True)
    ]
    
    # load dữ liệu chứa các thông tin node vào @nodes
    nodes = spark.read.csv("./data/transport-nodes.csv", 
                           header=True, 
                           schema=StructType(node_fields))
    
    # load dữ liệu chứa các thông tin về relationship vào @rels
    rels = spark.read.csv("./data/transport-relationships.csv",
                          header=True)
    
    """
    Vì bài toán của chúng ta là trên đồ thị vô hướng => ta cần thêm một đường
    đi ngược lại từ destination đến source.
    """
    reversed_rels = rels.withColumn("newSrc", rels.dst).\
                        withColumn("newDst", rels.src).\
                        drop("src", "dst").\
                        withColumnRenamed("newSrc", "src").\
                        withColumnRenamed("newDst", "dst").\
                        select("src", "dst", "relationship", "cost")
                        
    """
    Cuối cùng, ta cần kết hợp @rels và @reversed_rels lại thàng một
    dữ liệu chứa các cạnh duy nhất và lưu vào @relationships.
    """
    relationships = rels.union(reversed_rels)
    
    # trả về một GraphFrame
    return GraphFrame(nodes, relationships)
```

* Bây giờ, ta tạo một đồ thị và lưu vào biến `g` với hàm `createTransportGraph()`.

```python
g = createTransportGraph()
```

* Ta có thể thực hiện một vài phép lọc trên GraphFrame, mã dưới đây dùng để lọc ra các thành phố mà có dân số $x$ sao cho $100000 < x < 300000$. Sau đó sắp sếp tăng dân dựa trên **population**.

```python
g.vertices.filter("population > 100000 and population < 300000").\
    sort("population").show()
```

> Nhận xét:
> * Có hai thành phố là **Colchester** và **Ipswich** thỏa điều kiện trên.

* Tiếp theo, ta sẽ cho chạy thuật toán **Breadth First Search** với:
  * Đỉnh bắt đầu là thành phố có `id='Den Haag'`.
  * Đỉnh kết thúc là thành phố sao cho dân số $x$ của thành phố đó thỏa $100000 < x < 300000$ và thành phố này không phải là thành phố bắt đầu.

```python
from_expr = "id='Den Haag'" # đỉnh bắ đầu
to_expr = "population > 100000 and population < 300000 and id <> 'Den Haag'" # điều kiện cho đỉnh kết thúc
path = g.bfs(fromExpr=from_expr, toExpr=to_expr) # gọi thuật toán BFS tích hợp sẵn trong GraphFrames của Spark
```

* Ta có thể in ra kết quả bằng mã dưới đây.

```python
path.columns
```

> Nhận xét:
> * Hàm `.bfs()` trả về cho ta một mảng, trong đó `from` là đỉnh bắt đầu và `to` là đỉnh kết thúc.
> * Các phần tử có dạng `eX` ở đây là cạnh với id bằng `X` được GraphFrame tự định nghĩa khi ta tạo GraphFrame.
> * Các phần tử có dạng `vX` ở đây là các định với id bằng `X` được GraphFrame tự định nghĩa khi ta tạo GraphFrame. 

* Tuy nhiên, cách này không trực quan, ta hoàn toàn có thể chèn label vào như sau.

```python
cols = [c for c in path.columns if not c.startswith("e")]
p = path.select(cols).collect()[0]
paths = []
for c in cols:
    paths.append(p[c]['id'])
    print(paths[-1])
```

```python
plotGraph("./data/transport-relationships.csv", paths)
```

* Ngoài ra, Spark cũng cung cấp cho ta một hàm giúp ta tìm đường đi từ mọi đỉnh trong graph đến các đỉnh mà ta chỉ định trong hàm `shortestPaths()`. Dưới đây là mã giúp tìm kiếm mọi đường đi từ các đỉnh trong graph đến ba thành phố là **Colchester**, **Immingham**, **Hoek van Holland**.

```python
result = g.shortestPaths(["Colchester", "Immingham", "Hoek van Holland"])
result.sort(["id"]).select("id", "distances").show(truncate=False)
```
